```python
from bs4 import BeautifulSoup
import requests    
cdf1 = []
cdf2 = []

chanlist=["https://www.youtube.com/results?search_query=crime+channels",
          "https://www.youtube.com/results?search_query=sports+channels",
         "https://www.youtube.com/results?search_query=top+trending+channels"]

for links in chanlist:
    source = requests.get(links).text
    soup = BeautifulSoup(source, 'lxml')



    for content in soup.find_all('div', class_= "yt-lockup-content"):
        try:
            title = content.h3.a.text
            cdf1.append(title)

            description = content.find('div', class_="yt-lockup-description yt-ui-ellipsis yt-ui-ellipsis-2").text[:1000]
            cdf2.append(description)

        except Exception as e:
            description = None

        print('\n')
    cdf_dis = pd.DataFrame({'Channel Description': cdf2}).set_index("Channel Description")   
    cdf_title = pd.DataFrame({'Channel Name': cdf1}).set_index("Channel Name") 

cdf_dis

```
